```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('cost-revenue-clean.csv')
```

```python
data.describe()
```

```python
X = DataFrame(data,columns=['production_budget_usd'])
y = DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize = (10,6))
plt.scatter(X,y,alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

## Regression Theory
### Linear Regression

![title](LinearRegression.png)

y intercepts at 10
y increases by 1, when x increases by 2
so, the equation is like y=1/2x + 10 which is a slope

**y = mx + c**

![title](LR_Intercept.png)

Now, in ML the intercept and the slope is referred as Parameters

In this case, the c is considered as constant and slope coeffiecient (m). And also change the order like below. 

![title](ML_LineEquation.png)

**Fitted Value: The expected line value. The point on the line <br />
Residual: the difference between expected line value and actual value. The actual data value**
![title](ResidualFitted.png)

**Squaring the Residuals**
The sum of the Squared Residuals. This is the number the Linear Regression tries to minimize to choose the best parameters for the line
![title](SquaringTheResiduals.png)

In other words to find the best possible fit for our regression what we need to do is we need to choose an intercept that theta zero and we need to choose the slope theta one that minimizes the sum of the squared residuals and you will see this number also being referred to as the residual sum of squares or RSS
![title](ResidualSumOfSquares.png)

```python
regression = LinearRegression()
regression.fit(X,y)
```

Slope Coefficient:

```python
regression.coef_  # theta_1
```

```python
# Intercept
regression.intercept_ # theta_0
```

```python
plt.figure(figsize = (10,6))
plt.scatter(X,y,alpha=0.3)

plt.plot(X,regression.predict(X),color='red',linewidth=4)

plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

![title](Predict_50m.png)
![title](Predict_50mChart.png)

```python
regression.score(X,y) # Goodness of Fit: R Squared
```
